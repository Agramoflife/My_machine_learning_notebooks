## likelihood function 

You all ready know the likelihood function let's revisit it with example 

Certainly, let's keep it simple with another example:

**Example: Flipping a Coin**

Imagine you want to model how a fair coin behaves when you flip it. You have a hypothesis that the coin is fair, meaning it has an equal chance of landing heads or tails in each flip.

- **Model**: Your model is that the coin is fair, and there's a 50% chance of getting heads (H) and a 50% chance of getting tails (T) in each flip.

- **Data**: You start flipping the coin and record the results. For example, you might get a sequence like H, T, T, H, H, T, T.

- **Likelihood Function**: The likelihood function helps you see how well your model (the fair coin) matches your actual results. In this case, it should be high because you're assuming the coin is fair, and your recorded results are consistent with that assumption.

Here, you're not trying to "generate" the coin flips; you're observing them. The likelihood function is used to check how well your assumption (the coin is fair) matches the real-world data (the actual coin flips). If the likelihood is high, it means your model is a good fit for the data, suggesting that the coin is behaving as expected. If the likelihood is low, it might suggest that something is off, and the coin might not be fair.

In the coin-flipping scenario where you want to create a likelihood function, it's relatively straightforward because you're dealing with a simple binary outcome (heads or tails) and a basic model (fair coin with 50% chance for each outcome). The likelihood function can be constructed as follows:

1. **Model**: You assume that the coin is fair, which means that it has a 50% chance of landing heads (H) and a 50% chance of landing tails (T) in each flip.

2. **Data**: You have a sequence of coin flips, such as H, T, T, H, H, T, T.

3. **Likelihood Function (L(θ | x))**: The likelihood function represents how likely the observed data (x) is under different values of the parameter θ, which, in this case, is the probability of getting heads (H) in a single flip. Since you assume the coin is fair, θ is 0.5 (50%).

To create the likelihood function, you simply calculate the probability of observing your data sequence given the assumed θ value (0.5) for each flip, and then multiply those probabilities together because each flip is independent.

For example, if you want to calculate the likelihood of observing the sequence H, T, T, H, H, T, T assuming a fair coin (θ = 0.5):

L(0.5 | H, T, T, H, H, T, T) = P(H) * P(T) * P(T) * P(H) * P(H) * P(T) * P(T)

Since you assume a 50% chance for heads (H) and tails (T) in each flip:

L(0.5 | H, T, T, H, H, T, T) = (0.5) * (0.5) * (0.5) * (0.5) * (0.5) * (0.5) * (0.5) = 0.5^7 = 0.0078125

So, the likelihood of observing the sequence H, T, T, H, H, T, T assuming a fair coin is approximately 0.0078, or about 0.78%.

In practice, you would often work with the log-likelihood for computational reasons, which involves taking the natural logarithm of the likelihood, especially when dealing with many data points or more complex models.

A likelihood function, often denoted as L(θ | x), represents the probability of observing a particular set of data x given a specific set of parameter values θ in a statistical or probabilistic model. In simpler terms, it quantifies how well the model explains the observed data for different values of its parameters.

Here's a breakdown of the components:

- **L(θ | x)**: This notation signifies the likelihood function, where θ represents the parameter(s) of the model, and x represents the observed data.

- **θ**: Parameters are the values that you want to estimate or infer from the data. These parameters can be anything from the mean and variance of a normal distribution to coefficients in a regression model.

- **x**: Data represents the observed outcomes or measurements that you have collected or are working with. This can be a set of numbers, a sequence of events, or any other data that your model is supposed to explain or describe.

The likelihood function essentially tells you how probable it is to observe the given data, assuming that the model with parameter values θ is the true model. Mathematically, the likelihood function is often derived from the probability distribution of the model, and it is calculated by fixing the observed data and treating the parameters as variables.

The maximum likelihood estimation (MLE) is a common statistical method that aims to find the parameter values that maximize the likelihood function. In other words, MLE seeks the parameter values that make the observed data most probable under the assumed model.

In summary, a likelihood function is a fundamental concept in statistics and probabilistic modeling. It quantifies how well a model with specific parameter values explains the observed data, which is crucial for making statistical inferences and parameter estimation.

# stop words 
stop words are words that are commonly used in a language but are generally considered to be of little value in text analysis and information retrieval. These words are filtered out or ignored during text processing because they occur so frequently that they don't provide meaningful insights and can add noise to the analysis. Stop words include common words like "the," "and," "is," "in," "it," "an," "a," and so on.

```python
# just trying to see 
def text_process(x):
    msg = ''
    for i in x:
        if i not in string.punctuation:
            msg = msg + msg.join(i) + ''
    return msg
            
```

```python
message.message.apply(lambda x: text_process(x))
```

```python
def text_process_2(x):
    ''' This function removes the puctuation from the sentence and then removes the stop words'''
    
    # code for removing the punction 
    msg = ''
    for i in x:
        if i not in string.punctuation:
            msg = msg + msg.join(i) + ''
    
    # code for remove the stopwords 
    filtered_message = ' '
    for word in msg.split(' '):
        if word.lower() not in stopwords.words('english'):
            filtered_message = filtered_message + word + ' '
    
    return filtered_message.strip()
```

```python
message.message.apply(lambda x: text_process_2(x))
```

# fiting the model

```python
# we would basically do train test split first and then use the count vectorizer here
```

```python
from sklearn.model_selection import train_test_split
```

```python
msg_train,msg_test,label_train,label_test = train_test_split(message['tokenized_message'],message['label'],test_size = 0.2)
```

```python
msg_train.head()
```

```python
label_train.head()
```

```python
train_vectorized = vectorizer.transform(msg_train)
test_vectorized = vectorizer.transform(msg_test)
```

```python
train_vectorized 
```

```python
train_array = train_vectorized.toarray()
test_array = test_vectorized.toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
spam_detect_model = GaussianNB()
spam_detect_model.fit(train_array,label_train)
```

```python
train_preds = spam_detect_model.predict(train_array)
test_preds = spam_detect_model.predict(test_array)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(label_train,train_preds))
print("\n")
print(confusion_matrix(label_test,test_preds))
```

```python
# Print the classification report for train and test
print(classification_report(label_train,train_preds))
print("\n")
print(classification_report(label_test,test_preds))
```
