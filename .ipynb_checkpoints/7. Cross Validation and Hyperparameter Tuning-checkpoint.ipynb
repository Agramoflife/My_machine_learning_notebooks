# Variance inflation factor
It is used to find the multicolinearity between fetaures 

Now consider this there are deepu, anil, sachin and mehra now mehra has some special skills which no one can do, Anil and Sachin also have unique skill deepu has skills which sachin has and also mehra has so if you are a manager you would layoff deepu as his contribution can be outsourced to other members of the team mathematically we would say that deepu has a very high vif which means it's contribution to the work can be done by other members.

A high VIF indicates that a predictor variable is highly correlated with other predictor variables in the model, which can lead to unstable and unreliable coefficient estimates. This doesn't necessarily mean that the variable lacks special skills or unique contributions. It means that the variable's contribution to the model can be explained by other variables in the model.

```python
#Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
def vif_df(X):
    df = pd.DataFrame()
    df['Variables'] = X.columns
    df['VIF'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
    return df
```

```python
def vif_df(X):
    """ this will create a df of vif and the input it takes is a df"""
    #creating a empty df
    df = pd.DataFrame()
    # creating variable columns and naming it as the columns in the input data frame
    df['Variables'] = X.columns
    
    #creating vif column with the hlep of list of values 
    df['V.I.F'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
    
    return df
```

# Cross validation 

suppose you have a 10000 rows data 

If you have a dataset with 10,000 rows and you want to perform 5-fold cross-validation, here's how the process would work:

1. **Dataset Preparation:** Start with your dataset containing 10,000 rows.

2. **Partitioning into Folds:** Divide your dataset into 5 roughly equal-sized subsets or folds. Each fold should ideally contain 2,000 rows. These folds are typically created randomly or in a stratified manner (stratified sampling ensures that the class distribution remains consistent across folds if you're working on a classification problem).

3. **Iteration 1:** In the first iteration, you would use Fold 1 as the test set and Folds 2, 3, 4, and 5 as the training set. Train your machine learning model on the training data (8,000 rows) and evaluate its performance on the test data (2,000 rows). Record the performance metrics.

4. **Iteration 2:** In the second iteration, use Fold 2 as the test set and Folds 1, 3, 4, and 5 as the training set. Again, train your model on the training data (8,000 rows), evaluate its performance on the test data (2,000 rows), and record the performance metrics.

5. **Repeat for 5 Iterations:** Repeat this process for the remaining iterations, using a different fold as the test set in each iteration.

6. **Performance Evaluation:** After all 5 iterations are complete, you will have 5 sets of performance metrics (e.g., accuracy, precision, recall) corresponding to each iteration. These can be averaged to obtain an overall estimate of your model's performance.

The key idea is that each of the 5 folds acts as the test set once, while the remaining 4 folds are used for training. This process ensures that your model is tested on different subsets of the data, helping to obtain a more reliable estimate of its performance and its ability to generalize to unseen data.

Keep in mind that 5-fold cross-validation is just one of many variations. You can choose a different value of k (e.g., 10-fold cross-validation, leave-one-out cross-validation) depending on your specific needs and the size of your dataset.

#a * L1 + b * L2
#alpha = a + b and l1_ratio = a / (a + b)
The l1_ratio basically tells us how much weightage you want to give to L1 and L2. If you put l1_ratio to 0.5 and solve for a and b you will get a = b which means setting l1_ratio to 0.5 gives equal values of a and b and this a and b are basically the coefficients of L1 and L2 which indicates equal penalty will be applied to L1 and L2
