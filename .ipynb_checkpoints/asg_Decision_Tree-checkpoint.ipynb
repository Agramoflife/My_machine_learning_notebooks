The split value of a feature in a Decision Tree Regression (or Classification) is decided using a criterion that measures how well the split separates the data points and reduces the impurity or error in each resulting subset. The most common criteria for splitting are Mean Squared Error (MSE) for regression and Gini impurity, entropy, or misclassification error for classification. Let's focus on Mean Squared Error (MSE) for regression as an example:

1. **Calculate the Mean of the Target Variable in the Parent Node (Before Split):**
   - Calculate the mean (average) of the target variable values for all data points in the parent node. This mean represents the "center" of the data.

2. **Evaluate Split Candidates:**
   - For each feature, consider different threshold values as potential split points. These threshold values could be the unique values of the feature or a set of values in between the minimum and maximum values.

3. **Calculate the MSE for Each Split Candidate:**
   - For each potential split, divide the data into two subsets: the left subset (data points where the feature is less than or equal to the threshold) and the right subset (data points where the feature is greater than the threshold).
   - Calculate the MSE for each subset separately. MSE is calculated as the average of the squared differences between the actual target values and the mean target value within the subset.

4. **Select the Split with the Lowest MSE:**
   - Choose the feature and threshold that result in the lowest MSE. This split is considered the "best" split based on the MSE criterion.

The idea behind using MSE is to find a split that minimizes the spread or variance of the target variable within the resulting subsets. In other words, the goal is to make the subsets as "homogeneous" as possible in terms of the target variable values. A lower MSE indicates that the split does a better job of reducing the prediction error within each subset.

It's important to note that while MSE is a common criterion for regression, other criteria like Gini impurity or entropy are used for classification tasks. The choice of criterion depends on the nature of the problem (regression or classification) and the specific requirements of the application.

## And how is it decided on which feature to split on 
In a Decision Tree, the decision of making strategic splits heavily affects a tree’s accuracy. The decision criteria are different for classification and regression trees.

In the case of regression trees, the most common technique is to choose the split point that reduces the variance of the target variable in the resulting subsets. Variance is a measure of the variability or spread in a set of data. Mathematically, it is the average squared deviation from the mean score. We want to split the data in a way that minimizes the variance in each subset.

Here's how it works:

1. **Calculate Variance of Target Variable**: The first step is to calculate the variance of the target variable in our dataset.

2. **Calculate Variance for Each Split**: For each feature, we consider each value as a potential split and divide our dataset into two subsets based on this value. We then calculate the variance of the target variable for each subset.

3. **Calculate Weighted Variance**: The variance for each potential split is calculated as the weighted average of the variances of the two subsets. The weights are proportional to the sizes of the subsets.

4. **Choose Best Split**: The feature and split point that result in the lowest weighted variance are chosen for the split.

This process is repeated recursively, resulting in a decision tree where each node represents a decision (split point) that minimizes the variance and thus maximizes the accuracy of predictions.

```python
#MAPE
def mean_absolute_percentage_error(y_t, y_p):
    y_t, y_p = np.array(y_t), np.array(y_p)
    return np.mean(np.abs((y_t - y_p) / y_t)) * 100



mean_absolute_percentage_error(y_test,y_predicted)
```

```python
#Adjusted R2 train
ntr=len(X_train)
ptr=len(X_train.columns)
adj_rt2 = 1-(1-rt2)*(ntr-1)/(ntr-ptr-1)
print(adj_rt2,ntr,ptr)
```

```python
#Adjusted R2 test
n=len(X_test)
p=len(X_test.columns)
adj_r2 = 1-(1-r2)*(n-1)/(n-p-1)
print(adj_r2,n,p)
```
