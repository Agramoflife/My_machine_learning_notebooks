## Cleaning Name stopwords

```python
# Specify the file path
file_path = 'StopWords_Names.txt'
stop_words_name = []
# Open the file in read mode ('r')
with open(file_path, 'r') as file:
    # Read the entire content of the file
    text = file.read()
    #negative_list.append(text)

# Now, the 'text' variable contains the content of the text file
stop_words_name.append(text)
```

```python
len(stop_words_name[0].strip().split('\n'))
```

```python
stop_words_name_updated = []
for word in stop_words_name[0].strip().split('\n'):
    if len(word) >15:
        stop_words_name_updated.append(word.split(' ')[0])
    else:
        stop_words_name_updated.append(word)
        
```

```python
stop_words_name_updated = set(stop_words_name_updated)
```

## cleaning the currences stopwords

```python
# Specify the file path
file_path = 'StopWords_Currencies.txt'
stop_words_currencies = []
# Open the file in read mode ('r')
with open(file_path, 'r') as file:
    # Read the entire content of the file
    text = file.read()
    #negative_list.append(text)

# Now, the 'text' variable contains the content of the text file
stop_words_currencies.append(text)
```

```python
stop_words_currencies_updated = []
for word in stop_words_currencies[0].strip().split('\n'):
    word.split('|')
    for i in word.split('|'):
        stop_words_currencies_updated.append(i.strip())
```

```python
stop_words_currencies_updated = set(stop_words_currencies_updated)
```

```python
import string
from nltk.tokenize import wordpunct_tokenize

# Get the list of punctuation characters from Python's string module
punctuation = string.punctuation

# Tokenizing text based on punctuation using NLTK
text = "This is an example, showing how wordpunct_tokenize() works."
tokens = wordpunct_tokenize(text)

print("Punctuation characters:", punctuation)
print("Tokens:", tokens)
```

```python
dicty = {}
list_odd = []
list_even = []
for i in range(1,20):
    if i%2 == 0:
        list_even.append(i)
        dicty['even'] = list_even
    else:
        list_odd.append(i)
        dicty['odd'] = list_odd
        
print(dicty) 
```
