```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data, such as values from a normal distribution
mu = 0
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)  # Values along the x-axis
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)  # PDF values

# Create the PDF plot
plt.plot(x, pdf, label='PDF')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) Plot')
plt.legend()
plt.grid()

# Display the plot
plt.show()
```

```python
import os

# Define a list of video file extensions
video_extensions = ['.mp4', '.avi', '.mkv', '.mov', '.wmv', '.flv']

# List all files in the current directory
files_in_directory = os.listdir()

# Filter for video files
video_files = [file for file in files_in_directory if any(file.endswith(ext) for ext in video_extensions)]

# Check if there are any video files in the directory
if video_files:
    print("Video files found in the current directory:")
    for video_file in video_files:
        print(video_file)
else:
    print("No video files found in the current directory.")
```
