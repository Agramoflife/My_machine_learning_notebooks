```python
def gen(x):
    gen_list = []
    x = str(x)
    if x != 'nan':
        for i in eval(x):
            gen_list.append(i['name'])
        
    else:
        pass
    return gen_list
```

```python
df['genres_list'] = df['genres'].apply(lambda x: gen(x))
```

```python
imdb_df = df[['id','popularity','title','genres_list']]
```

```python
imdb_df.rename(columns = {'genres_list':'Genres'},inplace = True)
```
